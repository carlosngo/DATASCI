```python
import requests
import json
from bs4 import BeautifulSoup
```

```python
url = "https://newsapi.org/v2/everything"
payload = {
    'from': '2021-03-11',
    'to': '2021-03-12',
    'domains': 'inquirer.net',
    'apiKey': 'a57b432cf53a40aea2011db4929299f4'
}
```

```python
res = json.loads(requests.get(url, params=payload).text)
```

```python
articles = res['articles']
out_json = []
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.82 Safari/537.36'
}

for article in articles:
    author = article['author']
    title = article['title']
    date = article['publishedAt']
    a_url = article['url']
    page = requests.get(a_url, headers=headers)
    soup = BeautifulSoup(page.content, 'html.parser')
    page_content = soup.find('div',class_='entry-content')
    if page_content is not None:
        paragraphs = page_content.find_all('p')
        full_article = ''
        for paragraph in paragraphs:
            if len(paragraph.text.strip()) > 0:
                full_article = full_article + paragraph.text.strip() + '\n'
        out_json.append({
            'author': author,
            'title': title,
            'date': date,
            'content': full_article
        })
```

```python
with open('inquirer.json', 'w') as json_file:
    json.dump(out_json, json_file)
```
